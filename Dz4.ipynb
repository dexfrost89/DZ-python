Дан массив (100,2) координат, найти расстояние от каждой точки до каждой


```python
A = np.random.random((100, 2))
res = np.concatenate([np.subtract.outer(A[:, 0], A[:, 0]).reshape(100, 100, 1), np.subtract.outer(A[:, 1], A[:, 1]).reshape(100, 100, 1)], axis=2)
res = res ** 2
res = np.sqrt(np.sum(res, axis=2))
res.shape
res
```
